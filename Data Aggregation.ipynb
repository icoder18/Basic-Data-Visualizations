```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data/obama.csv',parse_dates=['year_month'])
# parse dates tells pandas to treat the column as a timestamp
# rather than just as regular data
data.head(10)
```

```python
plt.scatter(data.year_month,data.approve_percent,5,marker='o',alpha=0.3)
plt.show()
```

```python
data_mean=data.groupby('year_month').mean()
data_median=data.groupby('year_month').median()
```

```python
plt.plot(data_mean.index,data_mean.approve_percent,'red')
plt.plot(data_median.index,data_median.approve_percent,'green')
plt.legend(['Mean','Median'])
plt.scatter(data.year_month,data.approve_percent,3,marker='o',alpha=0.3)
plt.show()
```

```python
data=data[['year_month','approve_percent']]
```

```python
data_25=data.groupby('year_month').quantile(0.25)
data_75=data.groupby('year_month').quantile(0.75)
```

```python
plt.plot(data_75.index,data_75.approve_percent,'green')
plt.plot(data_mean.index,data_mean.approve_percent,'red')
plt.plot(data_25.index,data_25.approve_percent,'blue')
#plt.plot(data_median.index,data_median.approve_percent,'green')
plt.legend(['75 Percentile','Mean','25 Percentile'])
plt.scatter(data.year_month,data.approve_percent,3,marker='o',alpha=0.3)
plt.show()
```

```python
newdata=pd.read_csv('data/obama.csv',parse_dates=['year_month'])
newdata.head()
```

```python
plt.scatter(newdata.year_month,newdata.disapprove_percent,3,marker='x',alpha=0.3)
plt.show()
```

```python
newdata_mean=newdata.groupby('year_month').mean()
newdata_median=newdata.groupby('year_month').median()
```

```python
plt.scatter(newdata.year_month,newdata.disapprove_percent,3,marker='x',alpha=0.3)
plt.plot(newdata_mean.index,newdata_mean.disapprove_percent,color='red')
plt.plot(newdata_median.index,newdata_median.disapprove_percent,color='green')
plt.legend(['Mean','Median'])
plt.show()
```

```python
newdata=newdata[['year_month','disapprove_percent']]
```

```python
newdata_25=newdata.groupby('year_month').quantile(0.25)
newdata_75=newdata.groupby('year_month').quantile(0.75)
```

```python
plt.scatter(newdata.year_month,newdata.disapprove_percent,3,marker='x',alpha=0.3)
plt.plot(newdata_75.index,newdata_75.disapprove_percent,color='blue')
plt.plot(newdata_mean.index,newdata_mean.disapprove_percent,color='red')
plt.plot(newdata_25.index,newdata_25.disapprove_percent,color='green')
plt.legend(['75 Percentile','Mean','25 Percentile'])
plt.show()
```
