```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('data/countries.csv')
data.head()
```

```python
#Calculating the mean data of all countries
mean_data=data.groupby(['country']).mean()
mean_data.head()
```

```python
countries=set(data.country)
countries=sorted(countries)
print(type(countries))
print(len(countries))
```

```python
mean_data['country']=countries
mean_data.head()
```

#### Sorting Data Based on Population first, followed by GDP Per Capita and then Life Expectancy

```python

mean_data=mean_data.sort_values(by=['population','gdpPerCapita','lifeExpectancy'],ascending=False)
mean_data.head()
```

```python
for i in range(len(mean_data)):
    if mean_data.country.iloc[i]=='United States':
        print(i)
```

```python
resemblance=mean_data[0:5]
```

```python
resemblance.head()
```

```python
# As seen from the dataframe above, we cannot call any country similar to the US because of high variation in GDP.
# therefore, it'd be a better approach to sort on the basis of GDP Per Capita first
```

#### Sorting on the basis of GDP Per Capita followed by Population and then Life Expectancy

```python
mean_data=mean_data.sort_values(by=['gdpPerCapita','population','lifeExpectancy'],ascending=False)
mean_data.head()
```

```python
for i in range(len(mean_data)):
    if mean_data.country.iloc[i]=='United States':
        print(i)
```
