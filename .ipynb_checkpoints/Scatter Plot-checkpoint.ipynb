```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data/countries.csv')
data.head()
```

```python
data_2007=data[data.year==2007]
data_2007.head()
```

```python
plt.scatter(data_2007.gdpPerCapita,data_2007.lifeExpectancy,5)
#last parameter is size of points on the plot
plt.title("GDP vs Life Expectancy for 2007")
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
#Co-relation between the two columns
data_2007.gdpPerCapita.corr(data_2007.lifeExpectancy)
```

```python
np.log10([10,100,1000])
```

```python
plt.scatter(np.log10(data_2007.gdpPerCapita),data_2007.lifeExpectancy,5)
#last parameter is size of points on the plot
plt.title("GDP vs Life Expectancy for 2007")
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
#Co-relation between the two columns
np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)
```

```python
year=sorted(set(data.year))
```

```python
for y in year:
    data_year=data[data.year==y]
    plt.scatter(data_year.gdpPerCapita,data_year.lifeExpectancy,5)
    plt.title(y)
    plt.xlabel('GDP')
    plt.ylabel('Life Expectancy')
    plt.xlim(0,60000)
    plt.ylim(25,85)    
    plt.show()
    #plt.savefig(str(y),dpi=200)  #dots per inch, specifies the res of the img
    #plt.clf()
```

## Examine Relationship between GDP and Life Expectancy in 2007
* GDP is (GDP per Capita) x (Population)

```python
data_2007.head()
```

```python
gdp=data_2007.gdpPerCapita
d=gdp*data_2007.population/10**9
```

```python
plt.scatter(d,data_2007.lifeExpectancy,5)
plt.title('GDP vs Life Expectancy')
plt.xlabel('GDP (in billions)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
plt.scatter(np.log10(d),data_2007.lifeExpectancy,5)
plt.title('GDP vs Life Expectancy')
plt.xlabel('Log10 of GDP (in billions)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
print('Correlation between log10 of GDP and Life Expectancy:')
print(np.log10(d).corr(data_2007.lifeExpectancy))
print('Correlation between log10 of GDP Per Capita and Life Expectancy:')
print(np.log10(data_2007.gdpPerCapita/10**9).corr(data_2007.lifeExpectancy))
```
