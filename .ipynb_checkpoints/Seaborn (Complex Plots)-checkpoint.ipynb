```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.set_palette('Set2')
```

## Barplot

```python
#Give columns in x and y of the dataframe and 
#mention the dataframe in the data attribute

#Values are estimated using mean values unless otherwise mentioned
sns.barplot(x='sex',y='total_bill',data=tips)
```

## Countplot

```python
#Aggregation function is count always
#Plots the aggregates of the values in the columns
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

## Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#Shows the distibution of categorical data 
#Like on Thurday, min bill was the lower line (corresponding to 7)
#max bill was line corresponding to 30
#Maximum bill was distributed between the colored box shown
#outliers: represented by the dots outside the lines
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

## Violin Plot (similar to boxplot)

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## Distribution Plots

### dist plot

```python
sns.distplot(tips['total_bill'])
```

```python
#To remove KDE(Kernel Density Estimator) line 
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=20)
#Change no of bins
```

### KDE Graph

```python
#Plots just the kde line plot 
sns.kdeplot(tips['total_bill'])
```

### Joint plot

```python
#Used to join two plots
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#The right side histogram shows the tips if we used distplot on tips
#The top histogram shows the total_bill if we used distplot on total_bill
#The middle scatter plot represents the relation between the two 
#In this graph, the relation is almost linear
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

### If we want to compare all the features of a dataframe, we use 'pairplot'

```python
#Takes the entire dataframe as the parameter
sns.pairplot(tips)
```

```python
#On diagonal, we have the normal dist of the same var vs itself
```

```python
sns.pairplot(tips,hue='day')
```

```python
tips.head()
```

```python
#Correlation Function
#Takes all non-categorical variables and finds corr b/w them

tips.corr()
```

## Heat-Map

```python
#Accepts such kind of dataframes which have variables in both rows and columns
#Use corr mostly
corr=tips.corr()
```

```python
sns.heatmap(corr,annot=True)
```
