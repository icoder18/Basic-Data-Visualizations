```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[1,4,9])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Basic plot")
plt.show()
```

```python
plt.plot([1,2,3],[1,4,9],label="Squares")
plt.plot([1,2,3],[1,8,27],color="red",label="Cubes")
plt.xlabel("X-Values")
plt.ylabel("Y-Values")
plt.title("Basic plot of squares and cubes")
plt.legend()
#plt.show()
# To save the plot as an image
plt.savefig("exported_lineGraph")
```
